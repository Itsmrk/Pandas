# Pandas Concat() 

```python
import pandas as pd
```

```python
sr1 = pd.Series([1,2,3,4])
sr1
```

```python
sr2 = pd.Series([5,6,7,8])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
sr3 = pd.Series([5,6,7,8,9])
sr3
```

```python
pd.concat([sr3,sr2])
```

```python
pd.concat([sr1,sr2,sr3])
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
                  'Name':['A','B','C','D'],
                  'Class':[5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],
                  'Name':['E','F','G','H'],
                  'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2], axis = 1)
```

```python
pd.concat([df1,df2], axis = 0 , ignore_index = True)
```

```python
df1
```

```python
df2
```

```python
df3 = pd.DataFrame({'ID':[3,4],
                  'Name':['C','D'],
                  'Class':[7,8]})
df3
```

```python
pd.concat([df1,df3])
```

```python
pd.concat([df1,df3], axis = 1)
```

```python
pd.concat([df1,df3], axis = 1, join = 'inner')
```

```python
pd.concat([df1,df3], axis = 1, join_axes = [df1.index])
```

```python
pd.concat([df1,df3], axis = 1, join_axes = [df3.index])
```

```python
pd.concat([df1,df3.reindex(columns=df1.columns)])
```

```python
pd.concat([df1,df3.reindex_like(df1)])
```

```python
pd.concat([df1,df2], keys = ['df1','df2'])
```

```python
pd.concat([df1,df2], keys = ['First df','Second df'])
```

```python
pd.concat([df1,df2], keys = ['First df','Second df'], axis = 1)
```

```python
df4 = pd.DataFrame({'Marks':[100,102,103]})
df4
```

```python
pd.concat([df1,df4])
```

```python
pd.concat([df1,df4], sort = False)
```
