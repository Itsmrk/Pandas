```python
df = pd.read_csv("C:\\Users\IT's MRK\Desktop\Learning\Pandas\Book1.csv")
df
```

```python
df.dropna() # drops/removes all those row which have some NAN values 
```

```python
df.dropna(axis=1) # axis= 0 is default and use for rows , 1 is for columns
```

```python
df.dropna(how="any") 
```

```python
df.dropna(axis=0,how="all") # it drops all values and all columns when data frame has all null/NAN values 
```

```python
df.dropna(thresh = 7)
```

```python
df.dropna(subset = ['ID'])
```

```python
df.dropna(subset = ['Industry'])
```

```python
df.dropna(subset = ['Growth'])
```

```python
df.dropna(inplace = True)
df
```

```python
df = pd.read_csv("C:\\Users\IT's MRK\Desktop\Learning\Pandas\Book1.csv")
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Name':'none','Growth':0})
```

```python
df.fillna({'Name':'none','Growth':0 , 'Industry':"No Value"})
```

```python
df.fillna(method = 'ffill') # forward fill its is used to fill previous values to the NaN values 
```

```python
df.fillna(method = 'Pad') # forward fill its is used to fill previous values to the NaN values (its same as ffill)
```

```python
df.fillna(method = 'bfill') # backward fill its is used to fill backwarrd/bottom values to the NaN values (its same as ffill)
```

```python
df.fillna(method = 'ffill',axis = 1) # column wise forord fill
```

```python
df.fillna(method = 'bfill',axis = 1)
```

```python
df.fillna(method = 'ffill',axis = 0) # row wise forward fill
```

```python
df.fillna(0,limit = 1)
```

```python
df.fillna(0,limit = 2)
```

```python
df.fillna(0,limit = 3)
```

```python
df.fillna(method = 'ffill',limit = 2)
```

```python
df.fillna(5, inplace = True)
df
```
