# Pandas pivot_table() 
### The levels in the pivot table will b stored in MultiIndex objects (hierarchical indexes) on the index and columns of the result DataFrame
### Syntax : DataFrame.pivot_table()

```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\IT's MRK\Desktop\Learning\Pandas\movies_ratings.csv")
df
```

```python
df.pivot_table(index = 'Genre') # mean values 
```

```python
df.pivot_table(index = 'Genre', columns = 'Year')
```

```python
df.pivot_table(index = 'Year', columns = 'Ratings%')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'count')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'max')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'max',fill_value = 'none')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'max',fill_value = '0')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year',fill_value = '0',margins = True)
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'sum',fill_value = '0',margins = True)
```
