# Pandas Append() Function
### Used to append rows of other dataframe to the en o the given dataframe, returning a new dataframe object
### DataFrame.append()

```python
import pandas as pd 
```

```python
df1 = pd.DataFrame({'A':[1,2,3],
             'B':[10,20,30]})
df2 = pd.DataFrame({'A':[4,5,6],
             'B':[40,50,60]})
display(df1,df2)
```

```python
df1.append(df2)
```

```python
df1.append(df2,ignore_index = True)
```

```python
df2.append(df1,ignore_index = True)
```

```python
df1 = pd.DataFrame({'A':[1,2,3],
             'B':[10,20,30]})
df2 = pd.DataFrame({'c':[4,5,6],
             'B':[40,50,60]})
display(df1,df2)
```

```python
df1.append(df2,ignore_index = True)
```

```python
df1.append(df2,ignore_index = True,sort = False)
```
